```python
#  Correlation and Causation

#  correlation: measure of the extend of interdependence between variables
#  causation: relationship between causes and effect between two variables


#  Pearson Correlation
#  measures linear dependence between 2 variables X and Y

#  resulting coefficient is a value between -1 and 1 inclusive
#  1: total positive linear correlation
#  0: no linear correlation - two variables don't affect each other
# -1: total negative linear correlation 

# pearson correlation = df.corr()
```

```python
df.corr()
```

```python
#  want to know the significance of the correlation estimate 

#  p-value:
#  probability that the correlation between these 2 variables is statistically significant 

#  significance level of 0.05 = 95% confident that the correlation between the variables is significant 


#  < 0.001    strong evidence correlation is significant 
#  < 0.05     moderate 
#  < 0.1      weak 
#  > 0.1      no evidence that the correlation is significant 
```

```python
from scipy import stats
```

```python
#  pearson correlation coefficient and p-value of 


#  wheel-base vs price

pearson_coef, p_val = stats.pearsonr(df["wheel-base"], df["price"])

print("Pearson Correlation Coefficient is ", pearson_coef, " with P-Value of ", p_val)
```

```python
# p_val < 0.001, correlation between wheel-base and price is statistically significant
#  although the linear relationship isn't extrememly strong
```

```python
#  ANOVA   analysis of variance

#  test whether there are significant differences between the means of 2 or more groups


#  returns F-test score and p-value

#  F-test score: larger score means there is a larger difference between the means. assumption is theyre the same
#  p-value: how statistically significant is our calculated score value. 



#  if price is strongly correlated, expect sizeable F test score and small p-value


#  ANOVA analyzes the difference between different groups of the same variable

#  groupby function

# f_oneway to obtain f-test and pvalue
```

```python
#  see if different types "drive-wheels" impact price

grouped_test2 = df_gp_test[["drive-wheels", "price"]].groupby(["drive-wheels"])
grouped_test2.head()
```

```python
df_gp_test
```

```python
#  obtain the values of the method group 

grouped_test2.get_group("4wd")["price"]
```

```python
#  ANOVA

f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
#  F-test = strong correlation
#  p-value = imply almost certain statistical significance
```

```python
#  separately fwd and rwd

f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

```python
#  4wd and rwd

f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('rwd')['price'])  
   
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
#  4wd and fwd

f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('fwd')['price'])  
 
print("ANOVA results: F=", f_val, ", P =", p_val)   
```
